```python
## Sebastian Puerta-Hincapie
## Assignment #2
## CSC 301 - Numerical Issues
## Prof. Grimmelmann

# Libraries
import math;
import time as T;

# global variable definitions
PI = math.pi   # 3.141592
Euler = math.e # Precision value 2.718281

# Functions
# Naive Summation Algorithm
def naive_summation_algorithm1(s,n):
    for i in range(0,n):
        s += PI
    return s

# Compensated Summation Algorithm
def compensated_summation_algorithm1(s,e,n):
    # e - error, s - sum
        for i in range(0,n):
            temp = s
            y = PI + e
            s = temp + y
            e = (temp - s) + y
        s += e
        return s
    
# Naive Summation Algorithm using Euler for precision
def naive_summation_algorithm2(s,n):
    for i in range(0,n):
        s += PI
        s += Euler
    return s

# Compensated Summation Algorithm using Euler for precision
def compensated_summation_algorithm2(s,e,n):
    # e - error, s - sum
        for i in range(0,n):
            temp = s
            y = PI + e
            s = temp + y
            s += Euler
            y += Euler
            e = (temp - s) + y
        s += e
        return s

# Main program
def main():
    print("===== Start of Program =====")
    dash = '-' * 120
    # Testing Naive Summation
    print("Naive Summation")
    print(dash)
    print("Elapsed Time","%20s %20s %20s %20s %20s" % ("Time per Sum","Abs.Err","Rel.Err","Total Sum","Exact Value"))
    print(dash)
    for idx in range(6,10):
        N = pow(10,idx)
        #print("\nN = ", N)
        
        # Measuring Time of Algorithm
        start_time = T.time()
        sum_total = naive_summation_algorithm1(0,N)
        end_time = T.time()
        elapsed_time = T.time()
        time_per_sum = elapsed_time / (N - 1)
        
        # Calculations
        exact_value = (N * PI)
        abs_error = abs(exact_value - sum_total)
        rel_error = abs_error / exact_value
        
        # rounding
        sum_total = round(sum_total,4)
        elapsed_time = round(elapsed_time,3)
        time_per_sum = round(time_per_sum,4)
        exact_value = round(exact_value,4)
        abs_error = round(abs_error,4)
        rel_error = round(rel_error,4)
        
        # Outputs
        print(elapsed_time,"%20f %20f %20f %20f %20f" % (time_per_sum,abs_error,rel_error,sum_total,exact_value))
        
    # Testing Compensated Summation
    print("\nCompensated Summation")
    print(dash)
    print("Elapsed Time","%20s %20s %20s %20s %20s" % ("Time per Sum","Abs.Err","Rel.Err","Total Sum","Exact Value"))
    print(dash)
    idx = 0
    N = 0
    for idx in range(6,10):
        N = pow(10,idx)
        #print("\nN = ", N)
        
        # Measuring Time of Algorithm
        start_time = T.time()
        sum_total = compensated_summation_algorithm1(0,0,N)
        end_time = T.time()
        elapsed_time = T.time()
        time_per_sum = elapsed_time / (N - 1)
        
        # Calculations
        exact_value = (N * PI)
        abs_error = abs(exact_value - sum_total)
        rel_error = abs_error / exact_value
        
        # rounding
        sum_total = round(sum_total,4)
        elapsed_time = round(elapsed_time,3)
        time_per_sum = round(time_per_sum,4)
        exact_value = round(exact_value,4)
        abs_error = round(abs_error,4)
        rel_error = round(rel_error,4)
        
        # Outputs
        print(elapsed_time,"%20f %20f %20f %20f %20f" % (time_per_sum,abs_error,rel_error,sum_total,exact_value))
    
    # Testing Naive Summation w/ Euler Precision
    print("\nNaive Summation w/ Euler")
    print(dash)
    print("Elapsed Time","%20s %20s %20s %20s %20s" % ("Time per Sum","Abs.Err","Rel.Err","Total Sum","Exact Value"))
    print(dash)
    idx = 0
    N = 0
    for idx in range(6,10):
        N = pow(10,idx)
        #print("\nN = ", N)
        
        # Measuring Time of Algorithm
        start_time = T.time()
        sum_total = naive_summation_algorithm2(0,N)
        end_time = T.time()
        elapsed_time = T.time()
        time_per_sum = elapsed_time / (N - 1)
        
        # Calculations
        exact_value = (N * PI) + (N * Euler)
        abs_error = abs(exact_value - sum_total)
        rel_error = abs_error / exact_value
        
        # rounding
        sum_total = round(sum_total,4)
        elapsed_time = round(elapsed_time,3)
        time_per_sum = round(time_per_sum,4)
        exact_value = round(exact_value,4)
        abs_error = round(abs_error,4)
        rel_error = round(rel_error,4)
        
        # Outputs
        print(elapsed_time,"%20f %20f %20f %20f %20f" % (time_per_sum,abs_error,rel_error,sum_total,exact_value))
        
    # Testing Compensated Summation w/ Euler Precision
    print("\nCompensated Summation w/ Euler")
    print(dash)
    print("Elapsed Time","%20s %20s %20s %20s %20s" % ("Time per Sum","Abs.Err","Rel.Err","Total Sum","Exact Value"))
    print(dash)
    idx = 0
    N = 0
    for idx in range(6,10):
        N = pow(10,idx)
        #print("\nN = ", N)
        
        # Measuring Time of Algorithm
        start_time = T.time()
        sum_total = compensated_summation_algorithm2(0,0,N)
        end_time = T.time()
        elapsed_time = T.time()
        time_per_sum = elapsed_time / (N - 1)
        
        # Calculations
        exact_value = (N * PI) + (N * Euler)
        abs_error = abs(exact_value - sum_total)
        rel_error = abs_error / exact_value
        
        # rounding
        sum_total = round(sum_total,4)
        elapsed_time = round(elapsed_time,3)
        time_per_sum = round(time_per_sum,4)
        exact_value = round(exact_value,4)
        abs_error = round(abs_error,4)
        rel_error = round(rel_error,4)
        
        # Outputs
        print(elapsed_time,"%20f %20f %20f %20f %20f" % (time_per_sum,abs_error,rel_error,sum_total,exact_value))
    
    print("===== End of Program =====")
if __name__ == "__main__":
    main()
    
#### DISCUSSION #####
# For this exercise I ran four different functions, using the naive 
# compensated approaches. The first two functions do not use the Euler
# precision value. The following two functions use the Euler precision.
# I wanted to run different versions to see what the effect the Euler Precision
# value had on the summation results.

# The results are the following: 
# In all of the appraches as the power was increased so did the elapsed time.
# As the power value is increased by a power of 10
# the time per sum decreases by a power of 10 
# and the sum increases by a power of 10

# NAIVE APPROACH
# However, we can see a significant change in the absolute error from 
# 10 ^ 6 to 10 ^ 7 from a high to low but then as the power value goes
# from 10 ^ 8 to 10 ^ 9, it observed that the value goes up again. 

# COMPENSATED APPROACH
# In this case, we see a much more accurate result where the total sum
# and exact value match exactly. Therefore, there's no absolute error and
# no relative error. This is to be expected, since the compensated error
# can be much smaller since the algorithm involves computing errors in 
# the middle of each summation.
# According to wiki, "Kahan Summation Algorithm" significantly reduces
# the numerical error in the total obtained by adding a squence of finite
# precision floating point numbers, compared to the naive appoach. This
# done by keeping a separate running compensation - a variable to accumulate
# small errors.

# It can be concluded that there is indeed more error in the naive approach
# due to the difference between the exact value and calculate sum in comparison
# to the compensated approach which offers no error at all. In the end whether, 
# the euler precision value was added or not I was still able to see the error
# difference between the naive and compensated. Perhaps, adding Euler may add more
# accuracy to the calculations; however, when using the compensated summation approach
# the exact and total sum are still the same.
```
