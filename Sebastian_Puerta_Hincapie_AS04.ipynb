```python
## Sebastian Puerta-Hincapie
## Assignment #4 - cos(x) , d(cos(x))/dx - x = 1.0 rad
## CSC 301 - Numerical Issues
## Prof. Grimmelmann

# Libraries
import math;
import time;
import numpy;
from prettytable import PrettyTable
from texttable import Texttable

# global variables
ABSOLUTE = RELATIVE = FUNC = DERIV = 0
VALUE = 1.0
H = 0.001

# table set up
display_table =  PrettyTable()
display_table.field_names = ['Exact','Measured','Absolute Error','Relative Error']

# functions
# Ways to find cos(x)
def cosine_math():
    F = math.cos(math.degrees(VALUE))
    return F

def cosine_numpy():
    # initializing
    ABSOLUTE = RELATIVE = FUNC = 0
    display_table.clear_rows()
    
    # function
    FUNC = numpy.cos(numpy.degrees(VALUE))
    
    # Error
    ABSOLUTE = abs(FUNC - cosine_math())
    RELATIVE = ABSOLUTE / cosine_math()
    
    # display
    display_table.add_row([cosine_math(),FUNC,ABSOLUTE,RELATIVE])
    print(display_table)

def cosine_exp():
    # initializing
    ABSOLUTE = RELATIVE = FUNC = 0
    display_table.clear_rows()
    
    # function
    FUNC = (math.e**(VALUE*1j)).real
    
    # Error
    ABSOLUTE = abs(FUNC - cosine_math())
    RELATIVE = ABSOLUTE / cosine_math()
    
    # display
    display_table.add_row([cosine_math(),FUNC,ABSOLUTE,RELATIVE])
    print(display_table)

def cosine_taylor():
    # initializing
    n = 256
    x = 1
    ABSOLUTE = RELATIVE = FUNC = 0
    display_table.clear_rows()
    
    # function
    for i in range(0,n + 1):  
        summation = ((((-1)**i))*(x**(2*i))/(math.factorial((2*i))))   
        FUNC += summation

    # Error
    ABSOLUTE = abs(FUNC - cosine_math())
    RELATIVE = ABSOLUTE / cosine_math()
    
    # display
    display_table.add_row([cosine_math(),FUNC,ABSOLUTE,RELATIVE])
    print(display_table)

def cosine_in_terms_of_sine():
    # initializing
    ABSOLUTE = RELATIVE = FUNC = 0
    display_table.clear_rows()
    
    # function
    FUNC = math.sqrt(1 - math.sin(VALUE)**2)
    
    # Error
    ABSOLUTE = abs(FUNC - cosine_math())
    RELATIVE = ABSOLUTE / cosine_math()
    
    # display
    display_table.add_row([cosine_math(),FUNC,ABSOLUTE,RELATIVE])
    print(display_table)
    
# =========================================================================================

# Ways to find dcos(x)/dx
# derivative of cos(x) --> -sin(x)
def sine_math():
    dF = -math.sin(math.degrees(VALUE))
    return dF
# using numerical differentiation
def numerical_differentiation():
    # initializing
    ABSOLUTE = RELATIVE = DERIV = 0
    display_table.clear_rows()

    # Finding Derivative
    DERIV = ((math.cos(VALUE+H))- math.cos(VALUE))/H
    
    # Error
    ABSOLUTE = abs(DERIV - sine_math())
    RELATIVE = ABSOLUTE / sine_math()
    
    # display
    display_table.add_row([sine_math(),DERIV,ABSOLUTE,RELATIVE])
    print(display_table)

# using lagrange differentiation
def lagrange_differentiation():
    # initializing
    ABSOLUTE = RELATIVE = DERIV = 0
    display_table.clear_rows()
    
    # Finding Derivative
    x = math.degrees(VALUE)
    fx = math.cos(x)
    fxh = math.cos(x+H)
    f3e = -math.cos(x)
    DERIV = (((fxh- fx)/H)-((H/2)*f3e))
    
    # Error
    ABSOLUTE = abs(DERIV - sine_math())
    RELATIVE = ABSOLUTE / sine_math()
    
    # display
    display_table.add_row([sine_math(),DERIV,ABSOLUTE,RELATIVE])
    print(display_table)

# using forward/backward differentiation
def forward_backward_differentiation():
    # initializing
    ABSOLUTE = RELATIVE = DERIV = 0
    display_table.clear_rows()
    
    # FORWARD
    # Finding Derivative
    DERIV = (math.cos(VALUE + H) - math.cos(VALUE))/H
    
    # Error
    ABSOLUTE = abs(DERIV - sine_math())
    RELATIVE = ABSOLUTE / sine_math()
    
    display_table.add_row([sine_math(),DERIV,ABSOLUTE,RELATIVE])
    
    # BACKWARD
    # Finding Derivative
    DERIV = (math.cos(VALUE) - math.cos(VALUE - H))/H
    
    # Error
    ABSOLUTE = abs(DERIV - sine_math())
    RELATIVE = ABSOLUTE / sine_math()
    
    display_table.add_row([sine_math(),DERIV,ABSOLUTE,RELATIVE])
    
    # display
    print(display_table)

# using 3 point formulas
def three_point_differentiation():
    # initializing
    ABSOLUTE = RELATIVE = DERIV = 0
    display_table.clear_rows()
    
    # ENDPOINT
    # Finding Derivative
    x = math.degrees(VALUE)
    fx = math.cos(x)
    fxh = math.cos(x+H)
    fx2h = math.cos(x+2*H)
    f3e = -math.sin(x)
    DERIV = (((1/(2*H))*((-3*fx)+(4*fxh)-fx2h))+((H**2/3)*f3e))
    
    # Error
    ABSOLUTE = abs(DERIV - sine_math())
    RELATIVE = ABSOLUTE / sine_math()
    
    display_table.add_row([sine_math(),DERIV,ABSOLUTE,RELATIVE])
    
    # MIDPOINT
    # Finding Derivative
    x = math.degrees(VALUE)
    E = x-H
    fxh = math.cos(x+H)
    fhx = math.cos(x-H)
    f3e = -math.sin(E)
    DERIV = ((fxh-fhx)/(2*H))-((H**2/6)*f3e)

    # Error
    ABSOLUTE = abs(DERIV - sine_math())
    RELATIVE = ABSOLUTE / sine_math()
    
    display_table.add_row([sine_math(),DERIV,ABSOLUTE,RELATIVE])
    
    # display
    print(display_table)

# main program
def main():
    print("===== Start of Program =====")
    # Method 1
    print("\n===================================== COS(X) NUMPY ===========================================")
    cosine_numpy()
    # Method 2
    print("\n===================================== COS(X)  EXP ============================================")
    cosine_exp()
    # Method 3
    print("\n===================================== COS(X)  TAYLOR ==========================================")
    cosine_taylor()
    # Method 4
    print("\n===================================== COS(X)  COS IN TERMS OF SIN =============================")
    cosine_in_terms_of_sine()
  
    # Method 1
    print("\n\n===================================== dCOS(X)/dx  NUMERICAL =================================")
    numerical_differentiation()
    # Method 2
    print("\n===================================== dCOS(X)/dx  LANGRANGE ===================================")
    lagrange_differentiation()
    # Method 3
    print("\n===================================== dCOS(X)/dx  FORWARD/BACKWARD ============================")
    forward_backward_differentiation()
    # Method 4
    print("\n===================================== dCOS(X)/dx 3 POINT END/MID ==============================")
    three_point_differentiation()
    
    print("\n===== End of Program =====")
if __name__ == "__main__":
    main()
    
    
'''
As can be observed in the code above, four separate methods were used for finding cos(x) 
and the derivative of cos(x). In order to find the function, cos(x) where the x is 1.0 radian
the methods explored were the following: using the Numpy python library, using its respective exponential epression,
using taylor polynomials, and expresing cosine in terms of sines. By doing so, I was able to calculate results
that had a small error. It was interesting to see that regardless of the method to find cos(x),
the results obtained for each were very close - only off by 2-3 decimal points. Over the years, by taking for granted
that these values were equivalent, I was very pleased to finally see how closely accurate these expressions are.
Different expressions to obtain a close to equal result. In a way this is to be expected, if all these values were
rounded off, it could be concluded that the result is exact. However, observing the whole value itself, here we see
how equality may refer to "negligible error" meaning that the error is so small that it could be ignored,
hence, we have "equal" expressions. Fascinating how unprecise math really is - we just have the illusion of precision.

For the derivative of cosine, the methods of differentiation used were the following: Numerical, Langrange,
foward/back, and 3 point end/mid point. For this part, I did not have much of an expection - I was simply curious 
to see what would be the end results and how closely related these results would be. In the end it was gratifying, to 
obtain results with a small error and learn about alternative expressions to calculate the derivative of cosine. Overall,
more involved methods with more computation but still incredibly accurate to the "exact value".
'''
```
